```python
roman_emperor = pd.read_csv('roman-emperor-reigns.csv')
print(roman_emperor)
```

```python
death_class =dict(roman_emperor['Cause_of_Death'].sort_values().value_counts())
values = death_class.values()
keys = death_class.keys()

palette_color = sns.color_palette('bright')

plt.title("Death Cause Distribution Pie plot")

plt.rcParams["figure.figsize"] = (12,8)

plt.rcParams['font.size'] = 12


plt.pie(values, labels=keys, colors=palette_color, autopct='%.0f%%')
plt.show()
```
